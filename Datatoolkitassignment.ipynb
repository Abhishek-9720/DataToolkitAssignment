#2) How does broadcasting work in NumPy?
- Broadcasting in NumPy is a mechanism that allows arithmetic operations between arrays of different shapes by automatically expanding the smaller array’s shape to match the larger one, following specific rules.

###How Broadcasting Works
- **Shape Alignment:** When performing operations, NumPy compares array shapes starting from the rightmost dimension and works left. If the shapes are not the same, NumPy will virtually stretch dimensions of size 1 to match the other array, without actually copying data.

- **Compatibility Rules:** Two dimensions are compatible when:

  - They are equal, or

  - One of them is 1.

- **Dimension Padding:** If arrays differ in the number of dimensions, the smaller array is padded with ones on the left until both shapes have the same length.

**Example**

    import numpy as np

    a = np.array([[1, 2, 3], [4, 5, 6]])  # Shape (2, 3)
    b = np.array([10, 20, 30])            # Shape (3,)

    result = a + b
    # b is broadcast to shape (2, 3):
    # [[10, 20, 30],
    #  [10, 20, 30]]
    # Result:
    # [[11, 22, 33],
    #  [14, 25, 36]]
In this example, b is broadcast across the rows of a so that element-wise addition can occur.

- **Apply:** Functions (such as aggregation, transformation, or filtering) are applied independently to each group.

- **Combine:** The results from each group are combined into a new DataFrame or Series.

        'co2': [95, 90, 99, 104, 105, 94, 99, 104]
    })

#6)What are the differences between NumPy arrays and Python lists?
- Here are the key differences between NumPy arrays and Python lists :

- **Data Types**

  - Python lists can store elements of different types (heterogeneous).

  - NumPy arrays require all elements to be of the same type (homogeneous).

- **Memory Usage**

  - Python lists use more memory; each element is a separate Python object.

  - NumPy arrays are more memory-efficient, storing data in contiguous memory blocks.

- **Performance**

  - Python lists are slower for large-scale numerical operations.

  - NumPy arrays are much faster, especially for mathematical computations.

- **Element-wise Operations**

  - Python lists do not support element-wise operations natively; you need loops.

  - NumPy arrays support fast, vectorized element-wise operations (e.g., addition, multiplication).

- **Functionality**

  - Python lists are general-purpose and flexible for mixed data.

  - NumPy arrays are optimized for mathematical and scientific calculations.

- **Best Use Case**

  - Use Python lists when you need flexibility and mixed data types.

  - Use NumPy arrays when you need speed and efficiency for numerical data.

#8) What does the term “vectorized operation” mean in NumPy?
- A “vectorized operation” in NumPy refers to performing mathematical or logical operations on entire arrays at once, rather than processing each element individually in a loop. This is achieved by leveraging optimized, pre-compiled code (often written in C) that operates on sequences of data efficiently and quickly.

Key points about vectorized operations in NumPy:

- Operations (such as addition, subtraction, multiplication) are applied element-wise to arrays, eliminating the need for explicit Python loops.

- Vectorization results in concise, readable code and significant speed improvements, especially on large datasets.

- NumPy’s vectorized functions (like np.sum, np.add, or simply using +, -, * operators) internally use fast, low-level implementations for these operations.

- We can also vectorize custom functions using np.vectorize, allowing them to accept arrays as input and apply the function to each element automatically.

Example:

    import numpy as np
    a = np.array([1, 2, 3])
    b = np.array([4, 5, 6])
    result = a + b  # Adds each element: [5, 7, 9]
In summary, vectorized operations in NumPy are a core feature that enable fast, efficient, and expressive numerical computations on arrays without explicit loops.

#15)How does NumPy handle multidimensional arrays?
NumPy handles multidimensional arrays using its core data structure called the "ndarray", which is designed for efficient storage and manipulation of N-dimensional, homogeneous data. Here’s how NumPy manages these arrays:

**Creation of Multidimensional Arrays**
- We can create a multidimensional array by passing a nested list (or tuple) to np.array(). Each inner list represents a row (or a sub-array in higher dimensions), and the nesting depth determines the number of dimensions.

- NumPy also provides functions like np.zeros(), np.ones(), and np.arange() for creating arrays of specific shapes and values.

**Structure and Storage**
- The shape of a NumPy array is defined by a tuple of non-negative integers, where each integer specifies the size of the corresponding dimension.

- Internally, all data in a multidimensional array is stored in a contiguous block of memory as a 1D array. NumPy uses the shape and strides information to map multidimensional indices to this 1D memory layout.

**Indexing and Slicing**
- Elements are accessed using a comma-separated tuple of indices, e.g., arr[i, j] for 2D arrays, where i is the row and j is the column (zero-based indexing).

- Slicing allows you to extract subarrays, e.g., arr[0:2, 1:3] selects a block from a 2D array.

**Iteration**
- Iterating over a multidimensional array with a simple for loop iterates over the first axis (e.g., rows in a 2D array).

- To iterate over every scalar element, you need nested loops or can use np.nditer() for efficient iteration regardless of array dimensionality.

**Operations**
- NumPy provides vectorized operations and broadcasting, allowing you to perform element-wise mathematical operations, aggregations, and more across entire arrays without explicit Python loops.

- Matrix operations (dot products, transpositions, etc.) are supported natively for multidimensional arrays.

**Memory Layout**
NumPy uses row-major order (C-style), meaning the rightmost index changes fastest. This is different from some other languages (like MATLAB) that use column-major order.

#17) Explain the difference between apply() and map() in Pandas.
##Difference Between apply() and map() in Pandas
###Scope of Application

- map() is a method available only for Pandas Series. It applies a function or mapping (like a dictionary or another Series) to each element in the Series independently, returning a new Series with the transformed values.

- apply() is more versatile and is available for both Series and DataFrames. When used on a Series, it applies a function to each element (similar to map()). When used on a DataFrame, it can apply a function either row-wise or column-wise, allowing for more complex operations and aggregations.

###Element-wise vs. Row/Column-wise Operations

- map(): Strictly element-wise for Series—ideal for simple transformations or substitutions, such as mapping categorical values to numbers.

- apply():

  - On Series: Element-wise, similar to map(), but can handle more complex functions.

  - On DataFrames: Can operate along an axis (rows or columns), making it suitable for functions that require access to an entire row or column at once, such as aggregations or custom calculations.

**Examples**

    import pandas as pd

    # Example DataFrame
    df = pd.DataFrame({'A': [1, 2, 3], 'B': [10, 20, 30]})

    # map() on Series
    df['A'] = df['A'].map({1: 'one', 2: 'two', 3: 'three'})

    # apply() on Series
    df['B'] = df['B'].apply(lambda x: x / 10)

    # apply() on DataFrame (column-wise)
    df_sum = df.apply(lambda x: x.sum())
Summary

Use map() for simple, element-wise transformations on Series.

Use apply() for more complex operations—element-wise on Series, or along rows/columns in DataFrames, especially when the function requires access to more than one value at a time.

#21) Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy’s array slicing is faster than Python’s list slicing primarily due to differences in memory layout and data handling:

- **Contiguous Memory Storage:** NumPy arrays are stored in contiguous blocks of memory, which allows for efficient access and manipulation of data. This enables slicing operations to quickly create "views" of the original array without copying data, resulting in significant speed improvements.

- **Views vs. Copies:** When you slice a NumPy array, it typically returns a view—meaning no new data is copied, and both the original and sliced arrays share the same underlying data. In contrast, slicing a Python list always creates a new list and copies the data, which is slower, especially for large datasets.

- **Homogeneous Data Types:** NumPy arrays store elements of the same data type, making it possible to use optimized, low-level operations. Python lists can store mixed types, requiring more overhead to manage each element individually.

- **Optimized Implementation:** NumPy is implemented in C, leveraging highly optimized routines for memory access and data manipulation, further boosting slicing performance compared to native Python lists.

In summary, NumPy’s contiguous memory storage, use of views, homogeneous data types, and optimized implementation make its array slicing much faster than Python’s list slicing.

```python
#4)How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Select only numeric columns
numeric_diamonds = diamonds.select_dtypes(include='number')

# Calculate the correlation matrix for numeric columns
diamond_corr_matrix = numeric_diamonds.corr()

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(diamond_corr_matrix, cmap='coolwarm', annot=True)
plt.title('Diamonds Dataset Correlation Matrix Heatmap')
plt.show()
```

```python
#5) Generate a bar plot using Plotly.
import plotly.graph_objects as go

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Create a bar plot
fig = go.Figure(go.Bar(x=categories, y=values, marker_color='skyblue'))

# Add titles and labels
fig.update_layout(
    title='Sample Bar Plot',
    xaxis_title='Category',
    yaxis_title='Value'
)

# Show the plot
fig.show()
```

```python
#8) Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

# Data for three lines
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [5, 4, 3, 2, 1]

# Plotting multiple lines
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')

# Adding legend, labels, and title
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')

# Display the plot
plt.show()
```

```python
#9) Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 62, 90, 70]
}
df = pd.DataFrame(data)
print(df)
# Define the threshold
threshold = 75

# Filter rows where 'Score' is greater than the threshold
filtered_df = df[df['Score'] > threshold]

print(filtered_df)
```

```python
#10)Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

# Example: Load a sample dataset
df = sns.load_dataset("iris")

# Create a histogram for the 'sepal_length' column
sns.histplot(data=df, x="sepal_length")

plt.title("Distribution of Sepal Length in Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
#11) Perform matrix multiplication using NumPy.
import numpy as np

# Define two 2D arrays (matrices)
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
result = A @ B

print("Matrix Product of A and B is:\n", result)
```

```python
#13) Create a 3D scatter plot using Plotly.
import plotly.graph_objs as go
import numpy as np

# Sample data
z = np.linspace(0, 10, 50)
x = np.cos(z)
y = np.sin(z)

# Create a 3D scatter plot
trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=z,             # Color by z value
        colorscale='Viridis' # Color scale
    )
)

layout = go.Layout(title='3D Scatter Plot Example')

fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
